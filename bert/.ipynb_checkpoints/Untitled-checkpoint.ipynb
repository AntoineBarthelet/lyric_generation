```python
 lines = open('all_songs_with_metadata.txt', 'r').readlines()
```

```python
lines
```

```python
songs = []
current_song = []
start = False
for line in lines:
    if line == 'Lyrics:\n':
        # we are in a new song:
        current_song = []
        start = True
        continue
    if 'Song name:' in line:
        # we are have ended the previous song
        songs.append(current_song)
        start = False
    if line == '\n':
        continue
    if start:
        line = line.replace('\n', '').replace('-', '')
#         if line[-1] == ',':
#             list(line)[-1] = '.'
#             line = ''.join(line)
        current_song.append(line)
        
```

```python
import textdistance
from tqdm import tqdm
```

```python
dataset = []
distances = []
for song in tqdm(songs):
    if song == []:
        continue
    for i in range(int(len(song)/2)):
        l1 = song[i*2]
        l2 = song[i*2+1]
        d = textdistance.damerau_levenshtein(l1, l2)
        dataset.append([d, l1, l2])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame(dataset)
```

```python
df.sort_values(0).head(500)
```

```python
df.columns = ['distance', '#1 String', '#2 String']
```

```python
df = df.reset_index()
```

```python
df.columns = ['#1 ID', 'distance','#1 String', '#2 String']
df['#2 ID'] = df['#1 ID'] + 9704
```

```python
df['Quality'] = df['distance'].apply(lambda x: 1 if x>15 else 0)
```

```python
df = df[['Quality', '#1 ID', '#2 ID', '#1 String', '#2 String']]
```
